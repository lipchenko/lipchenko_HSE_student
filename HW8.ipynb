```python
with open('/content/pipeline_config.json', 'r', encoding='utf-8') as f:
  content = json.load(f)
 #Включила стемминг
content['steps']['stemming']['enabled'] = True
#Добавила numbers
content['steps']['stopwords']['custom_words'].append('numbers')
#Список шагов пайплайна
main_steps_in_pipe = {}
for key, value in content['steps'].items():
  if (value['enabled'] == True):
    main_steps_in_pipe[key] = value
print(main_steps_in_pipe)
#Изменила формат output_format
content['output_format'] = "vectors"
print(content)
#Создала упрощенную конфигурацию пайплана
simplified_content = content.copy()
simplified_content['steps'] = main_steps_in_pipe
print(simplified_content)
```

```python
with open('/content/nlp_aggregation.json', 'r', encoding='utf-8') as f:
  content = json.load(f)
#Нахожу с максимальным значением accuracy
model = max(content.items(),
            key = lambda item: item[1].get("accuracy", 0))
print(model)
#создаю список со значениями inference
inference = []
for key, value in content.items():
  it = value.get('inference_time')
  inference.append(it)

print(inference)
#среднее значение inference
mid_time = sum(inference)/len(inference)
print(mid_time)
#Создаю новый словарь со значениями accuracy и f1
metrics = {}
for key, value in content.items():
  ac = value.get("accuracy")
  fff = value.get("f1_score")
  metrics[key] = {"accuracy": ac,
                  "f1_score": fff}

print(metrics)
#Добавила новую модель

content["albert-base"] = {
    "accuracy": 0.87,
    "f1_score": 0.86,
    "inference_time": 55,
    "size_mb": 180
}
#создаю список маленьких моделей
mini_models = []
for key, value in content.items():
  sz = value.get('size_mb')
  if sz < 500:
    mini_models.append(key)

print(mini_models)

print(content)
```
