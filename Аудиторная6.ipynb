```python
orders = ("start", "stop", "pause", "quit")

while True:
  command = input("Введите команду: ") # Moved input inside the loop
  if command in orders:
    print(f"Выполняю {command}")

    if command == "quit":
        print("Выход из программы")
        break
  else:
    print("Неизвестная команда")
```

```python
shopping_list = []
print("добавить товар- 1, удалить товар - 2, показать список - 3, выйти - 4")

while True:
    user_command = input("Введите команду: ")

    if user_command == "добавить товар" or user_command == "1":
        good = input("Введите название товара: ")
        shopping_list.append(good)
        print(f"Товар '{good}' добавлен в список")

    elif user_command == "удалить товар" or user_command == "2":
        good = input("Введите название товара для удаления: ")
        if good in shopping_list:
            shopping_list.remove(good)
            print(f"Товар '{good}' удален из списка")
        else:
            print("Такого товара нет в списке")

    elif user_command == "показать" or user_command == "3":
            print(f"Ваш список покупок:{shopping_list}")


    elif user_command == "выйти" or user_command == "4":
        print("Выход из программы")
        break
    else:
        print("Неизвестная команда")
```

```python
email_set = set()
while True:
  email_address = input("Введите адрес ")
  email_set.add (email_address)
  if email_address == "":
    print(f"{sorted(email_set)}")
    break
```
