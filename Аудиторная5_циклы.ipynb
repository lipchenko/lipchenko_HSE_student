```python
mixed_data = [1, "hello", 3.14, [1, 2], 42, "world", 0, (5, 6), {"name" : "James"}, True, {7, 8, 9}]
first_list = []
second_list = []
third_list = []
fourth_list = []
just_list = []
float_list = []
dict_list = []
for item in mixed_data:
    if isinstance(item, int) and not isinstance(item, bool):
        first_list.append(item)
    elif isinstance(item, str):
        second_list.append(item)
    elif isinstance(item, list):
        third_list.append(item)
    elif isinstance(item, tuple):
        fourth_list.append(item)
    elif isinstance(item, bool):
        just_list.append(item)
    elif isinstance(item, float):
        float_list.append(item)
    elif isinstance(item, dict):
        dict_list.append(item)


print(first_list, second_list, third_list, fourth_list, just_list, float_list, dict_list)
```
