```python
numbers = []
while True:
  user_numbers = input()
  new_numbers = user_numbers.split()
  if any(number == '0' for number in new_numbers):
      break
  final_list = [int(number) for number in new_numbers
                  if int(number) > 0 and int(number) % 2 == 0]
  numbers.extend(final_list)
print(numbers, len(numbers))
```

```python
celsius = [0, 15, 25, 30, 40, 100]
fahrenheit = [i * 9/5 + 32 for i in celsius]
print(fahrenheit)
```

```python
grades = [45, 85, 92, 33, 67, 78, 90, 55, 29, 88]
passed = [i for i in grades if i >= 60]
print(passed)
```

```python
transactions = [100, -50, 200, -30, 150, -20, 300]
taxes = [i * 0.15 for i in transactions if i > 0]
print(taxes)
```

```python
products = ['футболка', 'кружка', 'блокнот']
colors = ['красный', 'синий', 'зеленый']
combinations = [(color, product)
                for color in colors
                for product in products]
print(combinations)
```

```python
employees = [('Иван', 45), ('Мария', 92), ('Петр', 33), ('Анна', 67)]


performance = ["отлично" if percent >= 90 else
               "хорошо" if percent >= 60 else
               'требует улучшения'
               for name, percent in employees]

print(performance)
```

```python
sales = [ {'name': 'Алексей', 'sales': 150, 'returns': 10},
         {'name': 'Ольга', 'sales': 200, 'returns': 5},
          {'name': 'Дмитрий', 'sales': 80, 'returns': 25},
           {'name': 'Елена', 'sales': 300, 'returns': 8} ]


top_managers = [person['name'] for person in sales
                if (person['sales'] - person['returns'] >= 150) and
                (person['returns'] / person['sales'] * 100 < 10)]
print(top_managers)
```
