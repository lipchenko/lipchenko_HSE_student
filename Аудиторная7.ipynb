```python
products = ["Ноутбук", "Мышь", "Клавиатура", "Монитор"]
prices = [50000, 2500, 4000, 30000]
quantities = [8, 45, 25, 12]

super_price = []

for i in range(len(products)):
  final_price = quantities[i] * prices[i]
  super_price.append(final_price)


max_value = max(super_price)

print(max_value)

```
